##normalizing X values##

```python
mean=x.mean(axis=0)
std=x.std(axis=0)
print(mean,std)
```

```python
X=((x-mean)/std)
```

```python
X
```

```python
## X and theta needs to be matrices
def sigmoid(X):
    return 1/(1+np.exp(-1*X))
def hypothesis(X,theta):
    return sigmoid(X*theta)
# Make sure Y is returned as a matrix of same size as that returned above
def gradient(X,y,theta):
    zz=hypothesis(X,theta)
    zz=zz-y
    thetam=X.T*zz
    #print(thetam,"theta")
    return (thetam/X.shape[0])
def gradientDescent(X,y,lr=0.4,iterations=10,batch_size=20):
    theta=np.zeros((X.shape[1],1))
    error_list=[]
    aa=0
    for i in np.arange(iterations):
        for j in np.arange(0,int(X.shape[0]/batch_size),1):
            theta=theta-lr*gradient(X[j*batch_size:(j+1)*batch_size,:],y[j*batch_size:(j+1)*batch_size,:],theta)
            aa=error_func(X,y,theta)
            #print(aa,"error")
            error_list.append(aa)
    return (theta,error_list)
def error_func(X,y,theta):
    y_=hypothesis(X,theta)
    #print(y_)
    #print(y_.shape,"y_shape")
    new=np.multiply(y,np.log2(y_))+np.multiply((1-y),np.log2(1-y_))
    print(-1*new.sum()/X.shape[0])
    return (-1*(new.sum()/X.shape[0]))
```

```python
X_log=np.mat(X.values)
on=np.ones((X_log.shape[0],1))
X_log=np.hstack((on,X_log))
print(X_log)
print(type(X_log))
Y_log=np.mat(y.values)
print(Y_log)
print(type(Y_log),Y_log.shape)
```

```python
(ans,error)=gradientDescent(X_log,Y_log)
ans
```

```python
f_ans=X_log*ans
print(f_ans)
print(f_ans.shape)
```

```python
def sigmoid(X):
    return 1/(1+np.exp(-1*X))
tt=sigmoid(f_ans)
print(tt)
print(tt.shape[0])
```

```python
test_check=test_check.astype('int')
```

```python
count=0
for i in range(int(test_check.shape[0])):
    if test_check[i,0]==Y_log[i,0]:
        count=count+1
print(count)
```
